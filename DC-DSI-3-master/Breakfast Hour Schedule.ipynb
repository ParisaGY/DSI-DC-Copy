# DSI-DC-3 Breakfast Hour



### DSI-DC-3 Breakfast Hour schedule.  
### Last Update: 2016/11/07


##### How to Manage DSI – 
Lecture Structured around what students can expect from the program as far as course-load and material as well as mental and emotional expectations.

##### Review - 
Time to review lab(s) or topics from earlier in the week.  Reviews will be two parted, 	with the first 30 minutes being peer-to-peer collaboration and the second 30 minutes 	composing of review of material with the instructors.

##### Review/Presentation –
Students will be tasked with creating a mini lecture on a topic we 	have covered in a previous week.  They will be responsible for constructing a 	presentation that not only shows their knowledge of the subject but also addresses a 	selected criterion based around technical interview questions.  Remaining time on 	these days will be devoted to lab or lecture review with the instructor.  

##### Portfolio/Blog – 
Time for students to create their weekly blog posts and/or update their 	portfolio.
	

##### Interview - 
Time will be used for interview preparation, including practice with Technical 	interview questions and coding challenges.  Mock interviews may also be incorporated 	in these time slots.

##### Visualization – 
Lectures and activities on proper visualization tactics and uses.  These activities and lectures will incorporate visualization using Matplot Lib, Seaborn and 	Tableau.  

##### Linear Algebra –
Series of short Lectures on Linear algebra done by Matt.
