```python
Non_Violent_Crimes=[ 'bad checks', 'bribery', 'drug/narcotic', 'drunkenness','embezzlement', 'forgery/counterfeiting', 'fraud', 'gambling', 'liquor,loitering', 'trespass']
non_criminal= ['non-criminal', 'runaway', 'secondary codes', 'suspicious occ', 'warrants'] 

Non_Violent_Crimes2 = []
non_criminal2=[]

for ii in range(len(Non_Violent_Crimes)):
    Non_Violent_Crimes2.append(Non_Violent_Crimes[ii].upper())

for ii in range(len(non_criminal)):
    non_criminal2.append(non_criminal[ii].upper())
```

```python
non_criminal2
```

```python
Non_Violent_Crimes2
```

```python
def s(x):
    for i in Non_Violent_Crimes2:
        if i==x:
            return 2
    for j in non_criminal2:
        if j==x:
            return 1
    else:
        return 0
sf_crime['value_feature'] = sf_crime['Category'].apply(s)
```
